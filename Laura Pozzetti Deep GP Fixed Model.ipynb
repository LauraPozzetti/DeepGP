```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.spatial.distance import cdist

from numpy.linalg import inv
```

```python
Xs = np.linspace(-5, 5, 100).reshape(-1,1)
X = np.random.uniform(-2, 2, 50).reshape(-1,1)
Y = X
```

```python
def f1(X):
    list = []
    for x in X:
        list.append(np.sin(x))
    return np.array(list)

def f2(X):
    list = []
    for x in X:
        list.append(x)
    return np.array(list)
```

```python
Z1 = X[0:30]
U1 = f1(Z1)

Z2 = U1
U2 = f2(Z2)
```

```python
m1 = U1
S1 = np.identity(30)

m2 = U2
S2 = np.identity(30)
```

```python
def K(x1, x2, l):
    K = np.exp(-cdist(x1,x2)**2/(l**2))
    return K

noise = 1*np.identity(len(X))
```

```python
Z1.shape
```

```python
f1_sample_mean = np.matmul(np.matmul(K(Xs,Z1,1),inv(K(Z1,Z1,1))),U1)
f1_sample_covariance = K(Xs,Xs,1)-np.matmul(np.matmul(K(Xs,Z1,1),inv(K(Z1,Z1,1))),K(Z1,Xs,1))+np.matmul(np.matmul(np.matmul(np.matmul(K(Xs,Z1,1),K(Z1,Z1,1)),S1),inv(K(Z1,Z1,1))),K(Z1,Xs,1))

f1_samples = np.random.multivariate_normal(f1_sample_mean.reshape(100,), f1_sample_covariance, 10)
```

```python
fig = plt.figure()
ax = plt.axes()
x = Xs
y = f1_samples.T
ax.plot(x,y,label='Layer f1')
```

```python
f2_sample_mean = np.matmul(np.matmul(K(Xs,Z2,1),inv(K(Z2,Z2,1))),U2)
f2_sample_covariance = K(Xs,Xs,1)-np.matmul(np.matmul(K(Xs,Z2,1),inv(K(Z2,Z2,1))),K(Z2,Xs,1))+np.matmul(np.matmul(np.matmul(np.matmul(K(Xs,Z2,1),K(Z2,Z2,1)),S2),inv(K(Z2,Z2,1))),K(Z2,Xs,1))

f2_samples = np.random.multivariate_normal(f2_sample_mean.reshape(100,), f2_sample_covariance, 10)

```

```python
fig = plt.figure()
ax = plt.axes()
x = Xs
y = f2_samples.T
ax.plot(x,y,label='Layer f2')
```
